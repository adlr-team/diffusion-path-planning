```python
import gym
import d4rl  # Import required to register environments, you may need to also import the submodule

# Create the environment
env = gym.make("maze2d-umaze-v1")

# d4rl abides by the OpenAI gym interface
env.reset()
env.step(env.action_space.sample())

# Each task is associated with a dataset
# dataset contains observations, actions, rewards, terminals, and infos
dataset = env.get_dataset()
# An N x dim_observation Numpy array of observations
print(dataset["observations"])

# Alternatively, use d4rl.qlearning_dataset which
# also adds next_observations.
```

```python
import requests
import h5py
import numpy as np
import pandas as pd


def download_hdf5(url, filename):
    response = requests.get(url)
    response.raise_for_status()  # Will raise an exception for bad responses
    with open(filename, "wb") as f:
        f.write(response.content)


# Example usage
url = "http://rail.eecs.berkeley.edu/datasets/offline_rl/maze2d/maze2d-open-sparse.hdf5"
filename = "maze2d-open-sparse.hdf5"
download_hdf5(url, filename)


def hdf5_to_dataframes(filename):
    with h5py.File(filename, "r") as file:
        dataframes = {}
        for key in file.keys():
            data = np.array(file[key])
            df = pd.DataFrame(data)
            dataframes[key] = df
    return dataframes


# Load data
dataframes = hdf5_to_dataframes(
    "/Users/magic-rabbit/Documents/ADLR/tum-adlr-ss24-18/maze2d-open-sparse.hdf5"
)

# Now you can use 'dataframes' which is a dictionary where each key is a dataset name and the value is a pandas DataFrame
```
